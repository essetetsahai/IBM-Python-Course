### Check for Missing Data

```python
#Replace '?' symbols with NaN
df = df.replace('?', np.NaN)
```

```python
#Identify missing values with Boolean values(T/F)
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Count missing values in each column using for loop
#value.counts() counts number of True values
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

### Dealing with Missing Values

#### 1. Replace missing values with column average

```python
#Replace missing normalized-losses with avg. normalized-losses

#Calculate mean of normalized-losses
avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized losses: ", avg_norm_loss)

#Replace NaN with Avg. 
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
df["normalized-losses"]
```

```python
# Replace missing values in the bore column with the column average

#Calculate avg. bore
avg_bore = df["bore"].astype('float').mean(axis=0)
print("Average of bore: ", avg_bore)


#Replace NaN with Avg.

df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
#Replace missing values in the stroke column with average stroke

#Calculate avg. stroke
avg_stroke = df["stroke"].astype('float').mean(axis=0)
print("Avg_stroke: ", avg_stroke)

#Replace NaN with Avg.
df["stroke"].replace(np.nan, avg_stroke)
```

```python
#Replace missing horsepower values with avg. horsepower

#Calculate avg. horsepower
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Avg_horsepower: ', avg_horsepower)

#Replace NaN with Avg. 
df["horsepower"].replace(np.nan, avg_horsepower, inplace = True)
```

```python
#Replace missing peak-rpm values with average peak-rpm

#Calculate avg. peak-rpm
avg_peakrpm = df["peak-rpm"].astype('float').mean(axis=0)
print("Avg_peakrpm: ", avg_peakrpm)

#Replace NaN with Avg. 
df["peak-rpm"].replace(np.nan, avg_peakrpm)
```

#### 2. Replace missing values with the most frequent value

```python
#Find all unique values and their counts in num-of-doors column

df["num-of-doors"].value_counts()
```

```python
#Or find the most frequent value(highest count) in the num-of-doors column

df["num-of-doors"].value_counts().idxmax()
```

```python
#Replace missing values in num-of-doors column with the most frequent value

df["num-of-doors"].replace(np.nan, "four", inplace = True)
```

#### 3. Drop rows that have missing data

```python
df = df.dropna(subset=["price"], axis =0)
df.head(5)
```

```python
# reset the index because of dropped rows

df.reset_index(drop=True)

df.head(10)
```

## Data Standardization

```python
#Convert mpg to L/100km for city-mpg
#L/100km = 235mi/gallon

df['city-L/100km'] = 235/df["city-mpg"]


df.head()
```

```python
#Transform mpg to L/100km in the column highway-mpg and rename it highway-L/100km

df["highway-mpg"] = 235/df["highway-mpg"]
df.head()

#Rename
df.rename(columns={"highway-mpg":'highway-L/100km'}, inplace=True)

#Check transformed data
df.head()
```

## Data Normalization

### Replace original value by original value/max value

```python
#Normalize length and width
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
#Normalize height
df['height'] = df['height']/df['height'].max()
```

# Statistical Analysis

```python
from scipy import stats
```

### Correlation

Wheelbase vs. Price

```python
#Calculate the Pearson Correlation Coefficient and P-value for the relationship between wheel-base and price

pearson_coef, p_value = stats.pearsonr(df['wheel-base'], df['price'])

print("Wheel-base vs. Price")
print("Pearson Correlation Coefficient: ", pearson_coef)
print("P-value: ", p_value)  
```

The p-value indicates the correlation between wheel-base and price is statistically significant.
The correlation coefficient indicates a positive, moderately strong linear relationship between wheel-base and price

Horsepower vs. Price

```python
#Calculate correlation coefficient and p-value of horsepower and price

pearson_coef, p_value = stats.pearsonr(df['horsepower'], df['price'])
print("Horsepower vs. Price")
print("Correlation Coefficient = ", pearson_coef)
print("P-value = ", p_value)
```

Horsepower and price have a strong, positive correlation, the correlation between the two variables is statistically significant.

City-mpg vs. Price

```python
pearson_coef, p_value = stats.pearsonr(df['city-mpg'], df['price'])
print("City-mpg vs. Price")
print("Correlation Coefficient = ", pearson_coef)
print("P-value = ", p_value)
```

City-mpg and price are negatively and moderately correlated. The correlation between city-mpg and price is statistically significant

### ANOVA

```python
Is there a statistically significant difference in pricing between drivetrains?
```

```python
Is there a statistically significant difference in pricing between drivetrains
```

```python
#Dataframe including only drive-wheels, body-style and price
df_one.head()
```

```python
#Group drive-wheels and price by drive wheels:

group_a = df_one[['drive-wheels', 'price']].groupby(['drive-wheels'])
```

```python
group_a.head(2)
```

```python
#Return the price for the group '4wd' 
group_a.get_group('4wd')['price']
```

```python
#F-test
f_val, p_val = stats.f_oneway(group_a.get_group('fwd')['price'], group_a.get_group('rwd')['price'], group_a.get_group('4wd')['price'])
print("F = ", f_val, "p-value = ", p_val)
```

There is a statistically significant difference in pricing between different drive-wheels. 

```python
#fwd and rwd
f_val, p_val = stats.f_oneway(group_a.get_group('fwd')['price'], group_a.get_group('rwd')['price'])
print("F = ", f_val, "p-value = ", p_val)
```

There is a significant price difference between fwd and rwd vehicles

```python
#4wd and rwd
f_val, p_val = stats.f_oneway(group_a.get_group('4wd')['price'], group_a.get_group('rwd')['price'])
print("F = ", f_val, "p-value = ", p_val)
```

There is a significant difference in pricing between 4wd and rwd  vehicles

```python
#fwd and 4wd
f_val, p_val = stats.f_oneway(group_a.get_group('fwd')['price'], group_a.get_group('4wd')['price'])
print("F = ", f_val, "p-value = ", p_val)
```

There is no significant difference in pricing between fwd and 4wd vehicles
