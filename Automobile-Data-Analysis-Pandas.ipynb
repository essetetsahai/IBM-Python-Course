```python
import pandas as pd
import numpy as np
```

```python
#Read the online file from url, and assign it to variable "df"
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
df = pd.read_csv(url,header = None)
```

```python
#Show the first 5 rows
print("The first 5 rows of the dataframe")
df.head(5)
```

```python
#Check bottom 10 rows of the dataframe
print("Bottom 10 rows")
df.tail(10)
```

Replace Header Names

```python
# create headers list
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
print("headers\n", headers)
```

```python
#Replace default headers with headers list
df.columns = headers
#Recheck top rows
df.head(10)
```

## Evaluating for Missing Data

```python
#Replace '?' symbols with NaN
df = df.replace('?', np.NaN)
```

```python
#Identify missing values with Boolean values(T/F)
missing_data = df.isnull()
missing_data.head(5)
```

```python
#Count missing values in each column using for loop
#value.counts() counts number of True values
for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

### Dealing with Missing Values

#### 1. Replace missing values with column average

```python
#Replace missing normalized-losses with avg. normalized-losses

#Calculate mean of normalized-losses
avg_norm_loss = df["normalized-losses"].astype('float').mean(axis=0)
print("Average of normalized losses: ", avg_norm_loss)

#Replace NaN with Avg. 
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace = True)
```

```python
df["normalized-losses"]
```

```python
# Replace missing values in the bore column with the column average

#Calculate avg. bore
avg_bore = df["bore"].astype('float').mean(axis=0)
print("Average of bore: ", avg_bore)


#Replace NaN with Avg.

df["bore"].replace(np.nan, avg_bore, inplace = True)
```

```python
#Replace missing values in the stroke column with average stroke

#Calculate avg. stroke
avg_stroke = df["stroke"].astype('float').mean(axis=0)
print("Avg_stroke: ", avg_stroke)

#Replace NaN with Avg.
df["stroke"].replace(np.nan, avg_stroke)
```

```python
#Replace missing horsepower values with avg. horsepower

#Calculate avg. horsepower
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print('Avg_horsepower: ', avg_horsepower)

#Replace NaN with Avg. 
df["horsepower"].replace(np.nan, avg_horsepower, inplace = True)
```

```python
#Replace missing peak-rpm values with average peak-rpm

#Calculate avg. peak-rpm
avg_peakrpm = df["peak-rpm"].astype('float').mean(axis=0)
print("Avg_peakrpm: ", avg_peakrpm)

#Replace NaN with Avg. 
df["peak-rpm"].replace(np.nan, avg_peakrpm)
```

#### 2. Replace missing values with the most frequent value

```python
#Find all unique values and their counts in num-of-doors column

df["num-of-doors"].value_counts()
```

```python
#Or find the most frequent value(highest count) in the num-of-doors column

df["num-of-doors"].value_counts().idxmax()
```

```python
#Replace missing values in num-of-doors column with the most frequent value

df["num-of-doors"].replace(np.nan, "four", inplace = True)
```

#### 3. Drop rows that have missing data

```python
df = df.dropna(subset=["price"], axis =0)
df.head(5)
```

```python
# reset index because of dropped rows
df.reset_index(drop=True)

df.head(20)
```

### Convert data types to proper format

```python
#Convert bore, stroke, price and peak-rpm to float

df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#Convert normalized losses to integer
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
```

```python
#List columns after conversion
df.dtypes
```

## Data Standardization

```python
#Convert mpg to L/100km for city-mpg
#L/100km = 235mi/gallon

df['city-L/100km'] = 235/df["city-mpg"]


df.head()
```

```python
#Transform mpg to L/100km in the column highway-mpg and rename it highway-L/100km

df["highway-mpg"] = 235/df["highway-mpg"]
df.head()

#Rename
df.rename(columns={"highway-mpg":'highway-L/100km'}, inplace=True)

#Check transformed data
df.head()
```

## Data Normalization

### Replace original value by original value/max value

```python
#Normalize length and width
df['length'] = df['length']/df['length'].max()
df['width'] = df['width']/df['width'].max()
```

```python
#Normalize height
df['height'] = df['height']/df['height'].max()
```

```python
#Show scaled columns

df[["length", "width", "height"]].head()
```
